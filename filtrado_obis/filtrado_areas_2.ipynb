This is a script that continues the mission of filtrado_areas. The objective of this script is to create an unique dataset that contains the number of genera, 'muestra' and observations.

First of all, we have to import the necessary libraries and dataset and create a unique dataset with all the data. One small detail is that we include an extra column with the correspondent filum. It follows the next correspondance:

cnidaria->1
crustacea->2
mollusca->3

```python
#We import the necessary libraries
import pandas
```

```python
#We import the necessary datasets
data1=pandas.read_csv('resultado_final_cnidaria_sinprofundidad.csv')
data2=pandas.read_csv('resultado_final_crustacea.csv')
data3=pandas.read_csv('resultado_final_mollusca_sinprofundidad.csv')
```

```python
#We create a new column in each dataset to identify the group
data1["filo"]=1
data2["filo"]=2
data3["filo"]=3
```

```python
#We concatenate the datasets into one
result=pandas.concat([data1, data2, data3])

#We see the result
print(result)
```

We have to create a unique 'muestra' for the three filum. So we delete the columns that are created in each dataset and we create the muestra and observations columns with the same procedure as in filtrado_areas

```python
#We delete the columns that are not needed
result1=result.drop(columns=['muestra','observaciones','fecha_hora_limpia'])
```

```python
#We create the column 'muestra'. There is a muestra for each combination of latitude, longitude and date
result1['muestra'] = result1.groupby(['decimalLatitude', 'decimalLongitude', 'fecha_hora']).ngroup()

#We check if we have done it correctly
print(result1)
```

```python
#We create the column 'observaciones' with the number of rows for each muestra
result1['observaciones'] = result1.groupby('muestra')['muestra'].transform('count')

#We reorder the columns by muestra for testing purposes
df_ordenado=result1.sort_values(by='muestra')
```

Finally, we save the table in a .csv file

```python
df_ordenado.to_csv('resultado_final_muestras.csv', sep=',', index=False)
```

Now, we have to create a new table with the number of genus, observations and 'muestra' of each of the 66 areas.

```python
#We create the dataset areas with the columns 'genusid', 'muestra', 'observaciones' and 'OBJECTID'. This
#are the columns that we need to create the dataset
areas=df_ordenado[["genusid","muestra", "observaciones","OBJECTID"]]
```

```python
#We create the column 'num_observations' with the number of rows for each OBJECTID
areas['num_observaciones'] = areas.groupby('OBJECTID')['OBJECTID'].transform('count')
```

```python
#We order by OBJECTID to facilitate the analysis
df_orden=areas.sort_values(by='OBJECTID')
```

```python
#We create the column 'num_muestra' with the number of different muestraa for each OBJECTID
num_muestra = areas.groupby('OBJECTID')['muestra'].nunique()
```

```python
#We create the column 'num_genero' with the number of different genera for each OBJECTID
num_genero=areas.groupby('OBJECTID')['genusid'].nunique()
```

```python
#We create a new dataframe merging the num_genero and num_muestra dataframes
resultado = pandas.merge(num_genero, num_muestra, on='OBJECTID')
```

```python
#We create the dataframe with the number of observations for each OBJECTID, we select the first one because it don't change on any OBJECTID
num_obs = areas.groupby('OBJECTID')['num_observaciones'].first().reset_index()
```

```python
#We merge to create the final dataframe with the number of observations, genera and muestra for each OBJECTID
resultado = resultado.merge(num_obs, on='OBJECTID')
```
